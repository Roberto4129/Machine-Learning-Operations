```python
#@app.get('/peliculas_duracion/{Pelicula}')
def peliculas_duracion(Pelicula: str):
    """
    Recibe el t√≠tulo de una pel√≠cula y devuelve la duraci√≥n y el a√±o de lanzamiento.
    """
    Pelicula = unicodedata.normalize('NFKD', Pelicula).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    pelicula = df_2.loc[df_2['title'].str.lower() == Pelicula.lower()]

    if pelicula.empty:
        return "No se encontr√≥ la pel√≠cula en el dataset"

    duracion = pelicula['runtime'].values[0]
    a√±o = pelicula['release_year'].values[0]

    return f"{Pelicula}. Duraci√≥n: {duracion} minutos. A√±o: {a√±o}"

```

def franquicia( Franquicia: str ): Se ingresa la franquicia, retornando la cantidad de peliculas, ganancia total y promedio
                    Ejemplo de retorno: La franquicia X posee X peliculas, una ganancia total de x y una ganancia promedio de xx

```python
#Obtener un nuevo conjunto de datos solo con las columnas necesarias
franquicias_pelis = df[["belongs_to_collection", "revenue"]]
#Normaliza la cadena a min√∫scula
franquicias_pelis["belongs_to_collection"] = franquicias_pelis["belongs_to_collection"].str.lower()
#exportar dataset
franquicias_pelis.to_csv("API1_fucnion3.csv",index=False)
```

```python
df_3 = pd.read_csv("API1_fucnion3.csv")
df_3
```

```python
def franquicia(Franquicia: str):
    franquicia = df_3[df_3['belongs_to_collection'].str.lower() == Franquicia.lower()]
    '''Se ingresa la franquicia, retornando la cantidad de peliculas, ganancia total y promedio'''
    if franquicia.empty:
        return "No se encontr√≥ la franquicia en el dataset"

    cantidad_peliculas = franquicia.shape[0]
    ganancia_total = franquicia['revenue'].sum()
    ganancia_promedio = franquicia['revenue'].mean()

    return {'franquicia':franquicia, 'cantidad':cantidad_peliculas, 'ganancia_total':ganancia_total, 'ganancia_promedio':ganancia_promedio}

    #return f"La franquicia {Franquicia} posee {cantidad_peliculas} pel√≠culas, una ganancia total de {ganancia_total} y una ganancia promedio de {ganancia_promedio}"
```

```python
#@app.get('/peliculas_pais/{Pais}')
def peliculas_pais(Pais: str):
    Pais = Pais.lower()
    """ Recibe el nombre de un pa√≠s y devuelve la cantidad de pel√≠culas producidas en el mismo."""
    df_clean = df_4.dropna(subset=['production_countries'])
    peliculas = df_clean[df_clean['production_countries'].str.contains(Pais, case=False)]
    cantidad_peliculas = len(peliculas)

    return {'pais': Pais, 'cantidad': cantidad_peliculas}
```

def get_director( nombre_director ): Se ingresa el nombre de un director que se encuentre dentro de un dataset debiendo devolver el √©xito del mismo medido a trav√©s del retorno. Adem√°s, deber√° devolver el nombre de cada pel√≠cula con la fecha de lanzamiento, retorno individual, costo y ganancia de la misma, en formato lista

```python
#Obtener un nuevo conjunto de datos solo con las columnas necesarias
director_pelis = df[['crew', 'title', 'release_date', 'return', 'budget', 'revenue']]
#Normaliza la cadena a min√∫scula
director_pelis['crew'] = director_pelis['crew'].str.lower()
director_pelis.dropna(subset=['crew'])
director_pelis.to_csv("API1_fucnion6.csv",index=False)#exportar dataset
```

```python
df_6 = pd.read_csv("API1_fucnion6.csv")
df_6
```

```python
#@app.get('/get_director/{nombre_director}')
def get_director(nombre_director: str):
    nombre_director = nombre_director.lower()
    nombre_director = unicodedata.normalize('NFKD', nombre_director).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    
    """  Ingesar  nombre de un director devuelve el √©xito del mismo medido a trav√©s del retorno. 
    nombre de cada pel√≠cula con la fecha de lanzamiento, retorno individual, costo y ganancia de la misma"""
    director_films = df_6[df_6['crew'].str.contains(nombre_director, case=False, regex=False) & ~df_6['crew'].str.contains('director', case=False, regex=False)]

    if director_films.empty:
        return "No se encontr√≥ informaci√≥n para el director: {}".format(nombre_director)
    
    cantidad_films = len(director_films)
    retorno_total = director_films['return'].sum()
    exito_director = retorno_total / cantidad_films

    detalles_peliculas = director_films[['title', 'release_date', 'return', 'budget', 'revenue']]
    detalles_peliculas = detalles_peliculas.rename(columns={
        'title': 'T√≠tulo',
        'release_date': 'Fecha de Lanzamiento',
        'return': 'Retorno',
        'budget': 'Presupuesto',
        'revenue': 'Ganancia'
    })

    return {'director': nombre_director, 'retorno_total_director': exito_director}, detalles_peliculas
   
 
```
