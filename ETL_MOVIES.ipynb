6. Algunos campos, como belongs_to_collection, production_companies y otros (ver diccionario de datos) estÃ¡n anidados, esto es o bien tienen un diccionario o una lista como valores en cada fila, Â¡deberÃ¡n desanidarlos para poder y unirlos al dataset de nuevo hacer alguna de las consultas de la API! O bien buscar la manera de acceder a esos datos sin desanidarlos.

```python
def convert_diccionario(diccionario):## Function to eval row datatype=dictionary and then extract it
    if isinstance(diccionario,str) and "{" in diccionario:
        dic=ast.literal_eval(diccionario)
        return dic["name"]
```

```python
movies_data["belongs_to_collection"]=movies_data["belongs_to_collection"].apply(convert_diccionario)## Apply dictionary function
```

```python
movies_data["belongs_to_collection"]
```

```python
#Production_companies => List
```

```python
def convert_lista(lista):## FunciÃ³n para evaluar la fila tipo de datos = lista y luego extraerla
    if isinstance(lista, str) and "{" in lista:
        L=[]
        for i in ast.literal_eval(lista):
            L.append(i["name"])
        return L
```

```python
movies_data["production_companies"]=movies_data["production_companies"].apply(convert_lista)## Apply list function
movies_data["genres"]=movies_data["genres"].apply(convert_lista)
movies_data["production_countries"]=movies_data["production_countries"].apply(convert_lista)
```

```python
movies_data["spoken_languages"]=movies_data["spoken_languages"].apply(convert_lista)
```
